```python
t = int(input())
for _ in range(t):
    J = str(input())
    D = str(input())
    J += 'z'
    print(J)
    D += 'z'
    print(D)
    j = 0
    d = 0
    S = ""
    while j < len(J) and d < len(D):
        if J[j:] < D[d:]:
            print(J[j:])
            print(D[d:])
            S += J[j]
            j += 1
        else:
            S += D[d]
            d += 1
    print("S before:", S)
    S = S[:-1]
    print("S After:", S)
    if j <len(J):
        S += J[j:-1]
    if d < len(D):
        S += D[d:-1]
    print(S)
```
